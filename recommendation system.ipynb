```python
import pandas as pd
import numpy as np
```

```python
ratings = pd.read_csv("Downloads/rating.csv",usecols=['user_id','anime_id'])
```

```python
anime = pd.read_csv("Downloads/anime.csv",usecols=['anime_id','name','genre','rating'])
```

```python
##here we merge the rating and anime data on "anime_id"

df=pd.merge(ratings,anime,on="anime_id")    
df.head()
```

```python
##dropping all the NaN values in the rows of 'name'
##creating a new variable where we will keep the no.of people rated a paticular anime.

combine_anime_rating=df.dropna(axis=0,subset=['name']) 
anime_rating_count=(combine_anime_rating.
        groupby(by=['name'])['rating'].           
        count().
        reset_index().
        rename(columns={'rating':'total_rating_count'})
        [['name','total_rating_count']]
        )
anime_rating_count.head()
```

```python
 ##now merging the anime rating count dataset with our prev. modified dataset!
rating_with_total_rating_count=combine_anime_rating.merge(anime_rating_count,left_on='name',right_on='name',how='left')
rating_with_total_rating_count.head()        
```

```python
##description
pd.set_option('display.float_format',lambda x:'%.3f'% x)      
print(anime_rating_count['total_rating_count'].describe())
```

```python
popularity_threshold=5000       ##from graph,i concluded a thresold value below which the rating count would not be considered!
rating_popular_anime=rating_with_total_rating_count.query('total_rating_count>=@popularity_threshold')  ##a moving with high rating but
rating_popular_anime.head()                                                                     ##low rating count may not be worthy to watch
```
